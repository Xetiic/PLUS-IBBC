```python
import re

# Regular expression pattern to extract the RPC values
pattern = r'(-?\d+(?:\.\d+)?)'

# Read the RPC values from the text file
with open('/root/PLUS-IBBC/data/rpc_01.txt', 'r') as file:
    rpc_values = file.read()

# Extract the RPC values using the regular expression
values = [float(x) for x in re.findall(pattern, rpc_values)]

# Create a dictionary with the keys and values
rpc_dict = {
    'LINE_OFF': values[0],
    'SAMP_OFF': values[1],
    'LAT_OFF': values[2],
    'LONG_OFF': values[3],
    'HEIGHT_OFF': values[4],
    'LINE_SCALE': values[5],
    'SAMP_SCALE': values[6],
    'LAT_SCALE': values[7],
    'LONG_SCALE': values[8],
    'HEIGHT_SCALE': values[9],
    'LINE_NUM_COEFF': values[10:30],
    'LINE_DEN_COEFF': values[30:50],
    'SAMP_NUM_COEFF': values[50:70],
    'SAMP_DEN_COEFF': values[70:]
}

print(rpc_dict)
```

```python
import rasterio
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Open the TIFF image and read the RPC metadata tags
with rasterio.open('/root/PLUS-IBBC/data/01.tif') as src:
    rpcs = rpc_dict

# Extract the RPCs from the dictionary
line_num_coeff = [float(x) for x in rpcs['LINE_NUM_COEFF']]
line_den_coeff = [float(x) for x in rpcs['LINE_DEN_COEFF']]
samp_num_coeff = [float(x) for x in rpcs['SAMP_NUM_COEFF']]
samp_den_coeff = [float(x) for x in rpcs['SAMP_DEN_COEFF']]
line_off = float(rpcs['LINE_OFF'])
samp_off = float(rpcs['SAMP_OFF'])
lat_off = float(rpcs['LAT_OFF'])
long_off = float(rpcs['LONG_OFF'])
height_off = float(rpcs['HEIGHT_OFF'])
line_scale = float(rpcs['LINE_SCALE'])
samp_scale = float(rpcs['SAMP_SCALE'])
lat_scale = float(rpcs['LAT_SCALE'])
long_scale = float(rpcs['LONG_SCALE'])
height_scale = float(rpcs['HEIGHT_SCALE'])

# Create an empty list to store the real-world coordinates
coords = []

# Loop through the pixels in the image
for row in range(src.height):
    for col in range(src.width):
        # Calculate the real-world coordinates for the pixel
        lat = sum(line_num_coeff[i] * row**(i+1) for i in range(20)) * line_scale + lat_off
        lon = sum(samp_num_coeff[i] * col**(i+1) for i in range(20)) * samp_scale + long_off
        h = (row - line_off) * line_scale + (col - samp_off) * samp_scale + height_off
        # Add the real-world coordinates to the list
        coords.append((lat, lon, h))


```
